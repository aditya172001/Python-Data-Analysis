# Data Analysis of Billionaires 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv",encoding = "ISO-8859-1")
df = pd.DataFrame(data)
df.head()
```

# Checking for Null Values

```python
print(df.isnull().sum())
```

```python
df = df.dropna() #Dropping the Null values
```

```python
print(df.isnull().sum()) #Cross checking for Null values
```

```python
df.head() #Printing the first 5 values after deleting the Null value rows
```

# Top 10 Billionaires according to their Net Worths

```python
df = data.sort_values(by = ["finalWorth"], ascending=False).head(10)
plt.figure(figsize=(20, 10))
sns.histplot(x="personName", hue="finalWorth", data=df)
plt.show()
```

# Top 5 Domains

```python
a = data["category"].value_counts().head()
index = a.index
sources = a.values
custom_colors = ["skyblue", "yellowgreen", 'tomato', "blue", "red"]
plt.figure(figsize=(5, 5))
plt.pie(sources, labels=index, colors=custom_colors)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Top 5 Industries to Become a Billionaire", fontsize=20)
plt.show()
```

# Top 5 Countries in terms of Billionares

```python
a = data["country"].value_counts().head()
index = a.index
Countries = a.values
custom_colors = ["skyblue", "yellowgreen", 'tomato', "blue", "red"]
plt.figure(figsize=(5, 5))
plt.pie(Countries, labels=index, colors=custom_colors)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Top 5 Countries with Most Number of Billionaires", fontsize=20)
plt.show()
```
